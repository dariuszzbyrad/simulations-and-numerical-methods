The model is a system of three ordinary differential equations now known as the Lorenz equations:

\begin{equation}
 \begin{cases}
    \frac{dx}{dt} = \sigma (y - x) \\
    \frac{dy}{dt} = x * (\rho - z) - y \\
    \frac{dz}{dt} = x * y - \beta * z
\end{cases}
\end{equation}

where: \\
$\sigma$ - Prandtl number \\
$\rho$ - Rayleigh number \\
$\beta$ - Expansion coefficient 

```python
# Imports libs
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Constant (defined by Lorenz)
sigma = 10
rho = 28
beta = 8/3.0

# Time space
t = np.linspace(0, 25, 10000)

# Initial conditions
xyz0 = [1.0, 1.0, 1.0]


def f(xyz, t, sigma, rho, beta):
  x, y, z = xyz # Unpack the state vector

  # Derivatives
  return [sigma * (y - x),
          x * (rho - z) - y,
          x * y - beta * z]

xyz = integrate.odeint(f, xyz0, t, args=(sigma, rho, beta))
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(xyz[:, 0], xyz[:, 1], xyz[:, 2])
plt.title('The dynamics for the Lorenz ODE')
plt.draw()
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, gridspec_kw={'height_ratios': [2, 2, 2]})

ax1.plot(t, xyz[:, 0])
ax1.set_ylabel('$x(t)$', fontsize=12)
ax1.set_xlabel('$t$', fontsize=12)

ax2.plot(t, xyz[:, 1])
ax2.set_ylabel('$y(t)$', fontsize=12)
ax2.set_xlabel('$t$', fontsize=12)

ax3.plot(t, xyz[:, 2])
ax3.set_ylabel('$z(t)$', fontsize=12)
ax3.set_xlabel('$t$', fontsize=12)

fig.tight_layout()
```
